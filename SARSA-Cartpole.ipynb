```python
import gym
import numpy as np
import random
from tqdm import tqdm
import matplotlib.pyplot as plt
from collections import defaultdict

# Initialize environment
env = gym.make('Taxi-v3')

# Discretize function for state space if needed
def discretize(state):
    return state  # Modify this if your state space needs discretization

# Function to choose action based on epsilon-greedy policy
def take_action(state, q_table, epsilon):
    if random.random() < epsilon:
        return env.action_space.sample()
    else:
        return np.argmax(q_table[state])

# Initialize hyperparameters
alpha_values = [0.01, 0.05, 0.1]
gamma_values = [0.9, 0.99]
epsilon_values = [0.1]
epsilon_decay = 0.99
min_epsilon = 0.01
episodes = 100

results = []

for alpha in tqdm(alpha_values, desc='alpha'):
    for gamma in tqdm(gamma_values, desc='gamma'):
        for epsilon in tqdm(epsilon_values, desc='epsilon'):
            q_table = defaultdict(lambda: np.zeros(env.action_space.n))
            rewards_per_episode = []
            x = []

            for episode in range(episodes):
                total_rewards = 0
                done = False
                state = discretize(env.reset()[0])

                while not done:
                    action = take_action(state, q_table, epsilon)
                    next_state, reward, done, _, _ = env.step(action)
                    total_rewards += reward

                    # Update Q-table
                    next_state = discretize(next_state)
                    best_next_action = np.max(q_table[next_state])
                    q_table[state][action] += alpha * (reward + gamma * best_next_action - q_table[state][action])
                    state = next_state

                rewards_per_episode.append(total_rewards)
                epsilon = max(min_epsilon, epsilon * epsilon_decay)
                x.append(episode)

            results.append((alpha, gamma, epsilon, x, rewards_per_episode))

# Plotting
plt.figure(figsize=(10, 6))
plt.grid(True)

for (alpha, gamma, epsilon, x, rewards) in results:
    plt.plot(x, rewards, label=f'α={alpha}, γ={gamma}, ε={epsilon}')

plt.xlabel('Episodes')
plt.ylabel('Reward per Episode')
plt.title('Q-Learning Performance')
plt.legend()
plt.show()
```
